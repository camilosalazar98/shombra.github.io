```python
#Create database and create Mentor table

import sqlite3

def printDB():
 
    try:
        result = c.execute("SELECT id, FName, LName, Company, Title, Interests, Email FROM Mentors")
 
        # You receive a list of lists that hold the result
        for row in result:
            print("ID :", row[0])
            print("FName :", row[1])
            print("LName :", row[2])
            print("Company :", row[3])
            print("Title :", row[4])
            print("Interests :", row[5])
            print("Email :", row[6])
 
    except sqlite3.OperationalError:
        print("The Table Doesn't Exist")
 
    except:
        print("Couldn't Retrieve Data From Database")
        


def dicts_factory(cursor, row):
    return dict((cursor.description[idx][0], value)
                for idx, value in enumerate(row))

#=================================

# connect() will open an SQLite database, or if it
# doesn't exist it will create it
db_conn = sqlite3.connect('SOMBRA.db')
 
print("Database Created")
 
# A cursor is used to traverse the records of a result
c = db_conn.cursor()

#=================================
try:
    db_conn.execute("CREATE TABLE Mentors(ID INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, FName TEXT NOT NULL, LName TEXT NOT NULL, Company TEXT NOT NULL, Title TEXT NOT NULL, Interests TEXT NOT NULL, Email TEXT NOT NULL);")
 
    db_conn.commit()
 
    print("Table Created")
 
except sqlite3.OperationalError:
    print("Table couldn't be Created")
#=================================


# the data to assign to those items

db_conn.execute("INSERT INTO Mentors (FName, LName, Company, Title, Interests, Email)"
                "VALUES ('Carl', 'Gordon', 'CUNY Building Performance Lab', 'Project Manager', 'HUGE ONE PIECE FAN', 'Carl@example.com')")
db_conn.execute("INSERT INTO Mentors (FName, LName, Company, Title, Interests, Email)"
                "VALUES ('Camilo', 'Salazar', 'Capital One', 'Software Engineer', 'Proud Brazilian Jiu-Jitsu White Belt', 'Camilo@example.com')")
db_conn.execute("INSERT INTO Mentors (FName, LName, Company, Title, Interests, Email)"
                "VALUES ('Alejandro', 'Morales', 'Stack Overflow', 'Data Scientist', '1v1 Me on FIFA 12', 'Alejandro@example.com')")
db_conn.execute("INSERT INTO Mentors (FName, LName, Company, Title, Interests, Email)"
                "VALUES ('Nathaly', 'Ardiles', 'Lockheed Martin', 'Test Engineer', 'Loves to Dance', 'Nathaly@example.com')")
db_conn.execute("INSERT INTO Mentors (FName, LName, Company, Title, Interests, Email)"
                "VALUES ('Miriam', 'Hernandez', 'General Assembly', 'Software Engineer', 'Loves to go swimming', 'Miriam@example.com')")
db_conn.execute("INSERT INTO Mentors (FName, LName, Company, Title, Interests, Email)"
                "VALUES ('Angel', 'Morales', 'Intel', 'Computer Engineer', 'Likes to play Fortnite Battle Royale', 'Angel@example.com')")
db_conn.commit()

print("Mentors Added")

db_conn.row_factory = dicts_factory

#=================================
```
